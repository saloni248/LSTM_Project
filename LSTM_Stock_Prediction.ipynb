```python
import numpy as np
import datetime
import pandas as pd

def sg39_str_to_datetime(sg39_s):
    sg39_split = sg39_s.split('-')
    sg39_year, sg39_month, sg39_day = int(sg39_split[0]), int(sg39_split[1]), int(sg39_split[2])
    return datetime.datetime(year=sg39_year, month=sg39_month, day=sg39_day)

def sg39_df_to_windowed_df(sg39_dataframe, sg39_first_date_str, sg39_last_date_str, sg39_n=3):
    sg39_first_date = sg39_str_to_datetime(sg39_first_date_str)
    sg39_last_date = sg39_str_to_datetime(sg39_last_date_str)

    sg39_dates = []
    sg39_X, sg39_Y = [], []

    sg39_current_date = sg39_first_date

    while sg39_current_date <= sg39_last_date:
        sg39_window = []
        sg39_temp_date = sg39_current_date

        # Find the values for the previous n valid trading days
        for i in range(sg39_n):
            sg39_temp_date -= datetime.timedelta(days=1)
            while sg39_temp_date >= sg39_dataframe.index[0]:
                sg39_temp_date_dt = pd.to_datetime(sg39_temp_date)
                if sg39_temp_date_dt in pd.to_datetime(sg39_dataframe.index):
                    sg39_window.insert(0, sg39_dataframe.loc[sg39_temp_date_dt])
                    break
                sg39_temp_date -= datetime.timedelta(days=1)
            else: #break out if we are out of dates.
                break

        # Check if we have enough data for the window
        if len(sg39_window) == sg39_n and pd.to_datetime(sg39_current_date) in pd.to_datetime(sg39_dataframe.index):
            sg39_X.append(sg39_window)
            sg39_Y.append(sg39_dataframe.loc[pd.to_datetime(sg39_current_date)])
            sg39_dates.append(sg39_current_date)

        sg39_current_date += datetime.timedelta(days=1)

    sg39_ret_df = pd.DataFrame({})
    sg39_ret_df['Target Date'] = sg39_dates

    sg39_X = np.array(sg39_X)
    for sg39_i in range(0, sg39_n):
        sg39_ret_df[f'Target-{sg39_n-sg39_i}'] = sg39_X[:, sg39_i]

    sg39_ret_df['Target'] = sg39_Y

    return sg39_ret_df
```

```python
import numpy as np

def sg39_windowed_df_to_date_X_y(sg39_windowed_dataframe):
    sg39_df_as_np = sg39_windowed_dataframe.to_numpy()

    sg39_dates = sg39_df_as_np[:, 0]

    # Exclude 'Target Date' and 'Unnamed: 0' column from the middle matrix
    sg39_middle_matrix = sg39_df_as_np[:, 2:-1]

    sg39_X = sg39_middle_matrix.reshape((len(sg39_dates), sg39_middle_matrix.shape[1], 1))

    # Selecting only the last column for sg39_Y
    sg39_Y = sg39_df_as_np[:, -1]

    return sg39_dates, sg39_X.astype(np.float32), sg39_Y.astype(np.float32)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a scaler object for Nifty
sg39_scaler_nifty = MinMaxScaler()

# Reshape the data for scaling (if necessary)
sg39_X_train_nifty_reshaped = sg39_X_train_nifty.reshape(-1, 1)
sg39_y_train_nifty_reshaped = sg39_y_train_nifty.reshape(-1, 1)

# Fit the scaler to the training data and transform it
sg39_X_train_nifty_scaled = sg39_scaler_nifty.fit_transform(sg39_X_train_nifty_reshaped).reshape(sg39_X_train_nifty.shape)
sg39_y_train_nifty_scaled = sg39_scaler_nifty.fit_transform(sg39_y_train_nifty_reshaped).reshape(sg39_y_train_nifty.shape)


# Scale validation and testing data using the fitted scaler
sg39_X_val_nifty_scaled = sg39_scaler_nifty.transform(sg39_X_val_nifty.reshape(-1, 1)).reshape(sg39_X_val_nifty.shape)
sg39_y_val_nifty_scaled = sg39_scaler_nifty.transform(sg39_y_val_nifty.reshape(-1, 1)).reshape(sg39_y_val_nifty.shape)

sg39_X_test_nifty_scaled = sg39_scaler_nifty.transform(sg39_X_test_nifty.reshape(-1, 1)).reshape(sg39_X_test_nifty.shape)
sg39_y_test_nifty_scaled = sg39_scaler_nifty.transform(sg39_y_test_nifty.reshape(-1, 1)).reshape(sg39_y_test_nifty.shape)
```

```python
# Model for Nifty
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau

sg39_model_nifty = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(128, return_sequences=True),  # Increased units, return sequences for stacked LSTM
    layers.LSTM(64),  # Added another LSTM layer
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),  # Added dropout for regularization
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# Learning rate scheduler
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

sg39_model_nifty.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

sg39_model_nifty.fit(sg39_X_train_nifty_scaled, sg39_y_train_nifty_scaled,
                    validation_data=(sg39_X_val_nifty_scaled, sg39_y_val_nifty_scaled),
                    epochs=100,
                    callbacks=[lr_scheduler])
```

```python
# Model for jpmc
sg39_model_jpmc = Sequential([layers.Input((3, 1)),
                          layers.LSTM(64),
                          layers.Dense(32, activation='relu'),
                          layers.Dense(32, activation='relu'),
                          layers.Dense(1)])

sg39_model_jpmc.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

sg39_model_jpmc.fit(sg39_X_train_jpmc, sg39_y_train_jpmc, validation_data=(sg39_X_val_jpmc, sg39_y_val_jpmc), epochs=100)
```
